# Questão 1: 
O valor de um computador deprecia exponencialmente com o tempo. Seu valor inicial é de R$ 4.000, e a taxa de depreciação anual é de 20%. Qual será o valor do computador após 5 anos? Use a fórmula:

$$
V(t) = V_0 \cdot (1 - r)^t
$$

onde $V_0$ é o valor inicial, $r$ é a taxa de depreciação e $t$ é o tempo em anos. Não se esqueça de traçar um gráfico mostrando o o valor do bem depreciado, ano a ano, nos próximos 15 anos.

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
```

$${V(t) = V_0 \times (1 - r)^t} $$

$${V(t) = 4000 \times (1 - 0.2)^{15}} $$

$${V(t) = 4000 \times (0.8)^{15}} $$

$${V(t) = 4000 \times 0.03518437208} $$

$${V(t) = 140.737488355} $$

```python
anos = np.linspace(0, 15, 256, endpoint = True)
preco = []

for t in anos:
  preco.append(4000 * ((1 - 0.2)**t))
plt.plot(anos, preco, color="blue", label=r'$f(x) = V0 \times (1 - r)^t$')
plt.xticks(np.arange(0, 16, 1))
plt.legend()
plt.grid()
plt.show()

print(preco[-1])
```

1. $\lim_{x \to 0} \frac{\sin(2x)}{x}$

2. $\lim_{x \to \infty} \frac{e^x}{x^3}$

1. 
$$ \lim_{x \to 0} \frac{\sin(2x)}{x} = 0 $$

$$ \lim_{x \to 0} \frac{\sin(2x)}{x} = 0 $$

$$ \lim_{x \to 0} \frac{2\cos(2x)}{1} = 0 $$

$$ \lim_{x \to 0} \frac{2\cos(2 \times 0)}{1} = 0 $$

$$ \lim_{x \to 0} \frac{2\times 1}{1} = 0 $$

$$ \lim_{x \to 0} = 2 $$


```python
x = np.linspace(0, 100, 256, endpoint = True)
y = []

for number in x:
  y.append(np.sin(2*number)/number)

plt.plot(x, y, color="blue")
plt.grid()
plt.show()
```

$$\lim_{x \to \infty} \frac{e^x}{x^3} = \infty$$
